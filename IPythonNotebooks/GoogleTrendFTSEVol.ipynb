```python
from pandas_datareader import data, wb
import datetime
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Get Index Data

```python
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2016,1,1)
FTSE100 = data.DataReader(name='^FTSE', data_source='yahoo', start=start, end=end)
FTSE100.describe()
```

```python
closing_data = pd.DataFrame()
closing_data['FTSE_Close'] = FTSE100['Close']
closing_data.describe()
```

```python
plt.figure()
closing_data.plot()
```
